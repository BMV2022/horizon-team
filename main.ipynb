```python
import pandas as pd
```

```python
predata = pd.read_csv("test_data.csv")
predata.head()
```

```python
predata.shape
```

```python
data = predata.sort_values(by="1",ascending=True)
data
```

```python
z=400
# 计算一部分的H
def calculateHValue(O,E,P,Ng=z):
    Og = O.count(1)
    Eg = E.count(1)
    PIg = (sum(P))/Ng
    print(Og,Eg,PIg)
    return (Og-Eg)*(Og-Eg)/(Ng*PIg*(1-PIg) + 0.001)
```

```python
data.shape[0]/z
```

```python
sumH = 0.0
for i in range(int(data.shape[0]/z)):
    caculate_data = data[i*z:i*z+z]
    sumH += calculateHValue(caculate_data["In-hospital_death"].to_list(),
        caculate_data["predictions"].to_list(),caculate_data["1"].to_list())
print("排序后的result：",sumH)
```

```python
sumH = 0.0
for i in range(int(data.shape[0]/z)):
    caculate_data = predata[i*z:i*z+z]
    sumH += calculateHValue(caculate_data["Survival"].to_list(),
        caculate_data["predictions"].to_list(),caculate_data["1"].to_list())
print("排序前的result：",sumH)
```
